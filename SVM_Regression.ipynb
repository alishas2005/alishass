```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR
```

```python
dataset = pd.read_csv("/content/level_salary.csv")
```

```python
dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
sns.scatterplot(x="Level",y="Salary",data=dataset)
plt.figure(figsize=(4,3))
```

```python
x_train, x_test, y_train, y_test = train_test_split(dataset.drop("Salary",axis=1),dataset["Salary"],test_size=0.2)
```

```python
x=dataset[["Level"]]
y=dataset["Salary"]
```

```python
sv = SVR(kernel="rbf")
sv.fit(x_train,y_train)
```

```python
sv.score(x_test, y_test)*100
```

```python
sv.score(x_train, y_train)*100
```

```python
sns.scatterplot(x="Level",y="Salary",data=dataset)

plt.plot(dataset["Level"],sv.predict(x),color="red")
```
